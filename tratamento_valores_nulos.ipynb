## Aqui vamos fazer o tratamento de dados para valores NaN / Null 

```python
# importar os pacotes
import pandas as pd
```

```python
# carregar o arquivo
df = pd.read_csv('dados.csv', delimiter=';')
df
```

```python
#vidualizando as primeiras linhas
df.head(10)
```

```python
#visualizaçao da dimensao
df.shape
```

```python
# verificando valores nulos
#isnull().sum() = soma qtd de valores nulos nas colunas
df.isnull().sum()
```

# Substituindo os valores:
## 1. Substituindo os valores nulos por um valor fixo

```python
# copia do DataFrame
df1 = df.copy()
df1.head(10)
```

```python
# substituir o valor nulo/NaN por um valor fixo
#pode inserir a médica valores 0 ou 1 ... 
# nesse caso vou substituir os valores por ZERO
#fillna = É um método do pandas usado para preencher valores ausentes
#inplcace=True para gravar a ação
df1.fillna(value= 0 , inplace= True)
df1.head(10)
```

## 2. Substituindo uma coluna especifica por um valor fixo

```python
df2 = df.copy()
df2.head(5)
```

```python
# Substuitindo os valores nulos/NaN da coluna QT_FILHOS
df2['QT_FILHOS'] = df2['QT_FILHOS'].fillna(value=0)
df2.head(10)
```

```python
# vericiar se tem valores nulos na coluna QT_FILHS
df2['QT_FILHOS'].isnull().sum()
```

## 3. Preencher com valores que vem antes ou depois do valor nulo/NaN

```python
#copiando o DataFrame
df3 = df.copy()
df3.head(5)
```

```python
# Preenchendo com as informaçoes da linha de frente e linha de tras

# ffill -> Front Fill -- Preenche os valores nulos com o registro ANTERIOR
df3['QT_FILHOS'] = df3['QT_FILHOS'].ffill()

# bfill -> Back Fill - Preenche os valores nulos com o registro seguinte 
df3['MESES_COMO_CLIENTE'] = df3['MESES_COMO_CLIENTE'].bfill()

df3.head(10)

# Na coluna QT_FILHOS na linha 3 o valor foi preenchido com valor que vem antes nesse caso o valor da linha anterior é 2
#Na coluna MESES_COMO_CLIENTE na linha 4 foi preenchido com o valor que vem após nesse caso o valor da linha 5 que é 36
```
